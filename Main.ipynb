## Data

The data used in this notebook is provided by kaggle (https://www.kaggle.com/tunguz/big-five-personality-test). The dataset includes information about how each participant answered each of the 50 questions. Each question contains a statement, and the participants answer with a 1 to 5, whether they do not agree with the statement or strongly agree with it. There is also data about how long each participant took to answer each question, measured in milliseconds. The dateload column represents the datetime stamp for when the page was loaded, screenw and screenh represent the screen resolution. Introelapse, testelapse, and endelapse are the time in seconds spent in the intro page, time it took to complete the test, and the time spent on the results page. Other information included is location of the country from where the test was completed, number of records from the same IP address, and the lattitude and longitude with high margin of error.

## Personality Correlations

Does being one personality mean that you are likely to be another one? In order to answer this question, we can do a Pearson's Correlation analysis between the scores of the 5 psychometric scores, and observe the results.

```python
import seaborn as sns
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
plt.figure(figsize=(12, 6))
sns.heatmap(filtered_countries[['E','A','N','C','O']].corr(),annot=True, cmap= cmap, vmin = 0, vmax= 0.5)
```

We need not comment on the diagonal of the plot, as that represents correlation of a column with itself, but although there appears to be mild correlations between some of the columns! 
- A high score of Extroverted is mildly correlated with a high score of Agreeable, and Neuroticism. 
- Agreeable has a weak correlation with Conscientusness, and a mild correlatio nwith Extroverted.
- Neuroticism has mild correlation with Extrverted and Counscientusness.
- Conscientousness is mildly correlated with Neuroticism, and weakly correlated with Agreeable.

The only trait with no correlations to any other trait is Open to Experience, which appears to be independent of the other psychometric scores, thus you never know who would be down to get weird, worth asking everyone as you might never know! (Friendly advice; do not ask your boss :))

## Testing Times

Are the psychometric scores correlated with test times? This is worth exploring as one personality might take longer to take the test, while others might just breeze by. <br>
A correlation analysis (again!) will show whether a person who scored high in a specific personality, might take longer to take the test. Before running the correlation analysis, it is worth removing outliers for times (for those who clicked like robots without reading, and for those who fell asleep during the test or left for lunch and came back after a week to it again). The function remove_outliers will help with this task.

```python
removed_outliers = remove_outliers(filtered_countries,
                                   column_names=['introelapse','testelapse','endelapse'])
plt.figure(figsize=(12, 6))
sns.heatmap(removed_outliers[['E','A','N','C','O', 'introelapse', 'testelapse','endelapse']].corr(),
            annot=True, cmap = cmap, fmt='.2f');
```

It is clear that the time it took to complete the test is not an indication of a certain personality trait, whether someone took a long time in the intro, to complete the survery or at the results page is not an indication of any personality trait.
